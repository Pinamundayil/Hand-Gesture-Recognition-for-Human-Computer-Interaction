```python
# import require libraries
import cv2
import mediapipe as mp
import pyautogui
```

```python
# Create a VideoCapture object to capture video from the default camera, the 0 reprecent the default webcam
cap = cv2.VideoCapture(0)
# Create a hand_detector object to detect hand landmarks
hand_detector = mp.solutions.hands.Hands()
#drawing_utils is used to draw the landmarks of the hand detected by the Hand Detection module on each frame of the video stream
drawing_utils = mp.solutions.drawing_utils
# To g et the size of the screen
screen_width, screen_height = pyautogui.size()
#Initialize index_y to 0
index_y = 0
while True:
    # Read a frame from the video capture object.
    # The cap.read() function reads a frame from the video capture object and stores it in the frame variable
    _, frame = cap.read()
    #flipping the frame horizontally before processing it
    #then landmarks are detected based on the assumption that the image is not mirrored, and the resulting hand gestures are more accurate
    frame = cv2.flip(frame, 1)
    frame_height, frame_width, _ = frame.shape
    #converting BGR into RGB
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    output = hand_detector.process(rgb_frame)
    hands = output.multi_hand_landmarks
    #print(hands)
    if hands:
        for hand in hands:
            drawing_utils.draw_landmarks(frame, hand)
            landmarks = hand.landmark
            for id, landmark in enumerate(landmarks):
                x = int(landmark.x*frame_width)
                y = int(landmark.y*frame_height)
                print(x, y)
                #Separate the index finger (landmark 8) so that we can use that as a mouse pointer
                if id == 8:
                    #draw a circle on the fra:me at the landmark's coordinates, and calculate the index finger's position on the screen in pixels
                    cv2.circle(img=frame, center=(x,y), radius=10, color=(0, 100, 233))
                    index_x = screen_width/frame_width*x
                    index_y = screen_height/frame_height*y
                    pyautogui.moveTo(index_x, index_y)
                if id == 4:
                    #draw a circle on the fra:me at the landmark's coordinates, and calculate the thumb finger's position on the screen in pixels
                    cv2.circle(img=frame, center=(x,y), radius=10, color=(0, 100, 210))
                    thumb_x = screen_width/frame_width*x
                    thumb_y = screen_height/frame_height*y
                    print('outside', abs(index_y - thumb_y))
                    #If the distance between the index finger's position and the thumb's position is less than 20 pixels, simulate a left mouse button click.
                    if abs(index_y - thumb_y) < 20:
                        pyautogui.click()
                        pyautogui.sleep(1)
                    #If the distance is less than 100 pixels, move the cursor to the position of the index finger on the screen.
                    elif abs(index_y - thumb_y) < 100:
                        pyautogui.moveTo(index_x, index_y)
    # Display the frame in a window
    cv2.imshow('Hand gesture recognition', frame)
    cv2.waitKey(1) 
    #The cv2.waitKey() function waits for a key event for the specified amount of time (in this case, 1 millisecond)
```
